# Scraping Stackoverflow Questions (test 2)

- Get question, views, vote-count, and question link

```python
import requests
from bs4 import BeautifulSoup
import json
```

# Question links

```python
questions_data = {
    "questions": []
}


for i in range(2):
    
    link = 'https://stackoverflow.com/questions?tab=frequent&page=' + str(i)
    
    print(link)
    res = requests.get(link)
    soup = BeautifulSoup(res.text, "html.parser")
    questions = soup.select(".question-summary")
    
    for que in questions:
        q = que.select_one('.question-hyperlink').getText()
        vote_count = que.select_one('.vote-count-post').getText()
        views = que.select_one('.views').attrs['title']
        questions_data['questions'].append({
            "question": q,
            "views": views,
            "vote_count": vote_count
        })
```

```python
questions_data = {
    "questions": []
}

link = 'https://stackoverflow.com/questions?tab=frequent&page=' + str(1)
    
res = requests.get(link)
soup = BeautifulSoup(res.text, "html.parser")
questions = soup.select(".question-summary")

for que in questions:
    q = que.select_one('.question-hyperlink').getText()
    question_link = 'https://stackoverflow.com'+ que.select_one('.question-hyperlink').attrs['href']
    vote_count = que.select_one('.vote-count-post').getText()
    views = que.select_one('.views').attrs['title']
    questions_data['questions'].append({
        "question": q,
        "views": views,
        "vote_count": vote_count,
        "question_link": question_link
    })
```
